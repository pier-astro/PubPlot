```python
import matplotlib.pyplot as plt
import numpy as np

pt = 1/72.27
jour_sizes = {"A&A": {"onecol": 256.*pt, "twocol": 523.5*pt, "titlesize": 12, "fontsize": 10, "ticksize": 9},
              "ApJ": {"onecol": 242.*pt},
             }

size_2 = jour_sizes["A&A"]["twocol"]
size = jour_sizes["A&A"]["onecol"]
golden = (1 + 5**0.5) / 2
plt.style.use("aanda.mplstyle")

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x) / 10  # Scale down to avoid large values
y4 = np.log1p(x)  # log(1+x) to avoid log(0)
y5 = np.sqrt(x) 
```

```python
# plotting
plt.figure(figsize=(size, size/golden))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='tan(x)/10')
plt.plot(x, y4, label='log(1+x)')
plt.plot(x, y5, label='sqrt(x)')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.xlabel('X-axis')
plt.ylabel('Y-axis (log scale)')
plt.title('We present, for the first time')
plt.legend()
plt.grid(True)
plt.show()
```
