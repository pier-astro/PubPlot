```python
import matplotlib.pyplot as plt
import numpy as np
from pubplot import setup_figure
from jbuilder import build_journal_sizes

jour_sizes = build_journal_sizes()


# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x) / 10  # Scale down to avoid large values
y4 = np.log1p(x)  # log(1+x) to avoid log(0)
y5 = np.sqrt(x) 
```

```python
# plotting
fig, ax = setup_figure(journal="aanda", columns=2, ncols=2, nrows=1)
plt.plot(x, y1, label='Godo')
plt.plot(x, y2, label='Godo')
plt.plot(x, y3, label='Godo')
plt.plot(x, y4, label='Godo')
plt.plot(x, y5, label='Godo')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.xlabel('Godo')
plt.ylabel('Godo (log scale)')
plt.title('Si gode')
plt.legend()
plt.show()
```

```python
from matplotlib.gridspec import GridSpec

fig = setup_figure(journal="aanda", columns=2, gridspec=True)
gs = GridSpec(nrows=2, ncols=2, figure=fig)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, :])
ax1.plot(x, y1, label='Godo 1')
ax2.plot(x, y2, label='Godo 2')
ax3.plot(x, y3, label='Godo 3')
for ax in [ax1, ax2, ax3]:
    ax.legend()
plt.show()
```
